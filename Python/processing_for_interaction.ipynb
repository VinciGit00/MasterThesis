## Subset

```python
df = df[df["IDStations"]==1266]
print(df.shape)
df.head()
```

```python

directions = ['N', 'S', 'E', 'O', 'NE', 'NO', 'SO', 'SE']

for direction in directions:
    df[f'wind_dir_10m_{direction}'] = (df['WE_mode_wind_direction_10m'] == direction).astype(int)

    df[f'wind_dir_100m_{direction}'] = (df['WE_mode_wind_direction_100m'] == direction).astype(int)

df = df.drop(columns=['WE_mode_wind_direction_10m', 'WE_mode_wind_direction_100m'])
```

```python
df = df.drop(columns=['AQ_pm25', 'AQ_co', 'AQ_so2'])
```

## Scaling

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler
scaler = MinMaxScaler()

# Specify columns to scale (excluding 'Time')
columns_to_scale = df.columns.difference(['Time'])

# Fit and transform the specified columns using MinMaxScaler
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# Create a 'Lockdown' column with values 1 for dates after '2020-03-09' and 0 otherwise
df['Lockdown'] = [1 if x > '2020-03-09' else 0 for x in df['Time']]

# Create a 'Season' column with values 1 for spring (March to May), 2 for summer, 3 for fall, and 4 for winter
df['Season'] = [1 if 3 <= int(x.split('-')[1]) <= 5 else
                2 if 6 <= int(x.split('-')[1]) <= 8 else
                3 if 9 <= int(x.split('-')[1]) <= 11 else 4 for x in df['Time']]

# Add a 'Station' column with values from 1 to 4 (assuming there are four stations)
df['Station'] = df['IDStations'].astype('category').cat.codes + 1

# Drop unnecessary columns
df = df.drop(columns=['IDStations', 'Latitude', 'Longitude', 'Time', 'Altitude'])

# Print the shape of the modified DataFrame
print(df.shape)

# Display the modified DataFrame
df
```
