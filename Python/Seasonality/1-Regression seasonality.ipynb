#Linear regression considering seasonality

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

# Crea i modelli per ogni mese
results = []
for month in range(1, 13):
    # Filtra i dati solo per il mese specificato
    df_month = df[df['Time'].str.contains(f'-{month:02d}-')]

    # Seleziona le colonne da usare come input
    X_cols = ['Latitude', 'Longitude', 'Altitude', 'WE_temp_2m', 'WE_wind_speed_10m_mean', 'WE_wind_speed_10m_max', 'WE_tot_precipitation', 'WE_precipitation_t', 'WE_surface_pressure', 'WE_solar_radiation', 'WE_rh_min', 'WE_rh_mean', 'WE_rh_max', 'WE_wind_speed_100m_max', 'WE_blh_layer_max', 'WE_blh_layer_min', 'EM_nh3_livestock_mm', 'EM_nh3_agr_soils', 'EM_nh3_agr_waste_burn', 'EM_nh3_sum', 'EM_nox_traffic', 'EM_nox_sum', 'EM_so2_sum', 'LI_pigs', 'LI_bovine', 'LA_hvi', 'LA_lvi', 'LA_land_use', 'LA_soil_use']
    X = df_month[X_cols].values

    # Seleziona la colonna da usare come output
    y = df_month['AQ_pm10'].values

    # Dividi il dataset in training set e test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Addestra il modello di regressione lineare
    reg = LinearRegression().fit(X_train, y_train)

    # Fai le previsioni sul training set e calcola le metriche
    y_train_pred = reg.predict(X_train)
    rmse_train = mean_squared_error(y_train, y_train_pred, squared=False)
    mae_train = mean_absolute_error(y_train, y_train_pred)
    r2_train = r2_score(y_train, y_train_pred)

    # Fai le previsioni sul test set e calcola le metriche
    y_test_pred = reg.predict(X_test)
    rmse_test = mean_squared_error(y_test, y_test_pred, squared=False)
    mae_test = mean_absolute_error(y_test, y_test_pred)
    r2_test = r2_score(y_test, y_test_pred)

    # Aggiungi le metriche alla lista dei risultati
    results.append({
        'Mese': month,
        'RMSE training set': rmse_train,
        'RMSE test set': rmse_test,
        'MAE training set': mae_train,
        'MAE test set': mae_test,
        'R2 training set': r2_train,
        'R2 test set': r2_test,
        'Coefficienti del modello': reg.coef_
    })

# Crea il dataframe dei risultati e stampalo come tabella
df_results = pd.DataFrame(results)
df_results = df_results.set_index('Mese')
print(df_results.to_string())
```
