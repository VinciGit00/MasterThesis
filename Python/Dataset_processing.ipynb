# Opening the dataframe

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Tesi/Agrimonia_Dataset_v_3_0_0.csv")
pd.set_option('display.max_columns', None)
print(df.shape)
df.head()
```

## Subset

```python
df = df[df["IDStations"]==1266]
print(df.shape)
df.head()
```

```python

directions = ['N', 'S', 'E', 'O', 'NE', 'NO', 'SO', 'SE']

for direction in directions:
    df[f'wind_dir_10m_{direction}'] = (df['WE_mode_wind_direction_10m'] == direction).astype(int)

    df[f'wind_dir_100m_{direction}'] = (df['WE_mode_wind_direction_100m'] == direction).astype(int)

df = df.drop(columns=['WE_mode_wind_direction_10m', 'WE_mode_wind_direction_100m'])
```

```python
df = df.drop(columns=['AQ_pm25', 'AQ_co', 'AQ_so2'])
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

columns_to_scale = df.columns.difference(['Time'])

scaler.fit(df[columns_to_scale])

df[columns_to_scale] = scaler.transform(df[columns_to_scale])

print(df.shape)
df
```
